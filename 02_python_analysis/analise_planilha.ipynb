```python
import pandas as pd
```

```python
# Leitura do Excel
df = pd.read_excel("data/Planilha_Vendas_Teste.xlsx", sheet_name="d_Pedidos")
```

```python
df
```

```python
print(df["Status"].value_counts())
```

```python
# Filtrar apenas pedidos com status "Concluida"
concluidos = df[df["Status"] == "Concluida"]
```

```python
concluidos
```

```python
# Somar a coluna de valor
total_valor = concluidos["Valor (R$)"].sum()
```

```python
print(f'Total de vendas concluídas: R${total_valor:.2f}')
```
